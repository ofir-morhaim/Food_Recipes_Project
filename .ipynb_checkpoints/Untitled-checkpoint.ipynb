```python
import numpy as np
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
%matplotlib inline
```

```python

df_Ingrediants = pd.read_csv("Ingredients.csv")
df_CleanTable = pd.read_csv("Final_clean_project_data_Recipe_table.csv")
df_range = pd.read_csv("Rating_categorial.csv")
```

```python
df_CleanTable
```

```python
X = df_CleanTable[["Preparation time in mins","Ingredients number"]].values
y = df_CleanTable["Rating_categorial"].values



XTrain, XTest, yTrain, yTest = train_test_split(X, y, random_state=1, test_size=0.25)
#XTrain= XTrain.reshape(-1, 1)
#yTrain= yTrain.reshape(-1, 1)
#XTest = XTest.reshape(-1, 1)
```

```python
gnb = GaussianNB()


# Train classifier
gnb.fit(XTrain,yTrain)

y_pred = gnb.predict(XTest)
y_pred_train = gnb.predict(XTrain)

# Print results
print('Accuracy on Train data= ', metrics.accuracy_score(y_true = yTrain, y_pred = y_pred_train))
print('Accuracy on test data= ', metrics.accuracy_score(y_true = yTest, y_pred = y_pred))
```
