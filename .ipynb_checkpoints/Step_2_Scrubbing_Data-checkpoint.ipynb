```python
def time_loc(loc):
    p_time_hr=0
    p_time_min=0
    try:
        if(loc.find("hr ")!=-1):
            loc = loc.replace("hr ",':',1)
        elif(loc.find("hrs ")!=-1):
            loc = loc.replace("hrs ",':',1)
        if(loc.find("min")!=-1):
            loc = loc.split("m")[0]
        if(loc.find(":")!=-1):
            p_time_hr = loc.split(":")[0]
            p_time_min = loc.split(":")[1]
        elif "hr" in loc:
            p_time_hr = loc.split("h")[0]
        else:
            p_time_min = loc.split("m")[0]
        total_time = (int(p_time_hr) * 60) + int(p_time_min)
        return total_time
    except:
        return None

def rating_loc(loc):
    rating = None
    try:
        rating = loc.split(":")[1]    
        rating = rating.split("%")[0]
        rating = float(rating)
        rating = int(rating *100)
        rating = float(rating)/100
        return float(rating)
    except:
        return None
    
    
def list_loc(loc):
    try:
        loc = loc.replace("['","").replace("']","").replace("',",",").replace(" '","")
        return loc
    except:
        return None
def calories_loc(loc):
    try:
        loc = loc.split(":")[1]
        loc = loc.split('\n')[0]
        loc = loc.replace(" ","")
    except:
        return None

    c=float(loc)
    return int(c)
```

```python
df = pd.read_csv("project_data_Recipe_table.csv")
try:
    df=df.drop(df[df["Ingredients number"]<2].index)
    df.drop(['Unnamed: 0'],axis='columns',inplace=True)
except:
    pass
```

```python
def count_duplicatives(df, col_name=None):
    if col_name==None:
        return df.duplicated().sum()
    else:
        return df.duplicated([col_name]).sum()
```

```python
dup_rows =  count_duplicatives(df)
print("The number of duplicate rows is " ,dup_rows)


print(" \nCount total NaN at each column in a DataFrame : \n\n",
      df.isnull().sum())


# Count total NaN at each row in a DataFrame
for i in range(len(df.index)) :
    temp= df.iloc[i].isnull().sum()
    if temp!=0:
        print(" Total NaN in row", i + 1, ":",temp)
        
df.drop_duplicates()
df = df.dropna()
```

```python


for index, row in df.iterrows():
    df['Preparation time in mins'][index]= time_loc(df["Preparation time in mins"][index])
    df['Rating'][index]= rating_loc(df["Rating"][index])
    df['Ingredients'][index]=list_loc(df['Ingredients'][index])
    df["Calories per serving"][index]= calories_loc(df["Calories per serving"][index])

df = df.loc[(df["Rating"]!=0) & (df["Rank number"]>20) &(df["Rating"]>=90)]
```
