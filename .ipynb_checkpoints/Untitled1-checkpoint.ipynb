```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt      
import math
import seaborn as sns
from collections import Counter
from sklearn.decomposition import PCA
```

```python

df_CleanTable = pd.read_csv("Final_clean_project_data_Recipe_table.csv")
df_Rating_Preparation=df_CleanTable[["Name","Rating","Rank number","Preparation time in mins_categorial"]].copy()

count = 81
last=0
number =1
array_rating=[]
df_Rating_Preparation.insert(len(df_Rating_Preparation.columns), "Rating_categorial", 0, True)
temp = 1
while int(temp) > 0:
    data={}
    df_Rating_Preparation.loc[(df_CleanTable["Rating"] < count) & (df_CleanTable["Rating"] >=last), "Rating_categorial"] = int(number)
    data["Start Range"] = last
    data["End Range"] = count-1
    data["Number"] = number
    array_rating.append(data)
    last=count
    count = count + 5
    number=int(number)+1    
    temp = df_Rating_Preparation["Rating_categorial"].isin([0]).sum()

df_range = pd.DataFrame(array_rating)
df_range


        


```
